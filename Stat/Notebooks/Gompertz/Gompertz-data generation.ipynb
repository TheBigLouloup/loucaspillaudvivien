# Hazard rate plot and parameter estimation in the Gompertz--Makeham model

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Dataset generation

In this section, we describe how the data contained in `my_sample` has been generated.

### Parameters, survival function and hazard rate

We first set the values of the true parameters $\lambda$, $\alpha$, $\beta$.

```python
true_lambda = .205
true_alpha = .788
true_beta = .432
```

The survival function $\mathbb{P}(T>t)$ is defined as follows.

```python
# Survival function
def true_survival(t):
  return np.exp(-(true_lambda*t + true_alpha/true_beta*(np.exp(true_beta*t)-1)))

# Grid of values of t
t_max = 5
n_tt = 10000
tt = np.linspace(0,t_max,n_tt)

# Plot the survival function
plt.plot(tt,true_survival(tt))
```

We may also visualise the hazard rate function $\lambda + \alpha \exp(\beta t)$.

```python
plt.plot(tt, true_lambda+true_alpha*np.exp(true_beta*tt))
```

### Data generation

We use the inverse CDF method to generate $T$ under the Gompertz distribution, by setting $T=F^{-1}(U)$ where $U$ is a uniform random variable on $[0,1]$ and $F$ is the CDF of $T$.

However the theoretical computation of $F^{-1}$ is not easy, so we proceed by discretisation.

```python
# Vector of the values of F(t) for t in tt
vect_of_cdf = 1-true_survival(tt)

# Inverse CDF
def inverse_cdf(u):
  i = np.searchsorted(vect_of_cdf, u, side='right') # Largest index such that vect_of_cdf[i] <= u
  return tt[i]
```

We now generate the sample.

```python
from scipy.stats import uniform

# Size of the sample
n_sample = 100000

# Set the seed for reproducibility
np.random.seed(0)

# Draw n_sample uniform variables
vect_of_unif = uniform.rvs(size=n_sample)

# Define the sample
the_sample = np.zeros(n_sample)
for i in range(n_sample):
  the_sample[i] = inverse_cdf(vect_of_unif[i])

# Export the sample to dataset.csv
np.savetxt('gompertz-dataset.csv', the_sample, delimiter=',')
```

We compare the histogram of the sample with the theoretical density of the Gompertz law.

```python
plt.hist(the_sample, density=True, bins=100)
plt.plot(tt, (true_lambda + true_alpha*np.exp(true_beta*tt))*true_survival(tt), color='r')
```

## Dataset

The array `my_sample` contains $n$ realisations $T_1, \ldots, T_n$.

```python
# Sample
my_sample = np.loadtxt('gompertz-dataset.csv', delimiter=',')

# Size of the sample
n = np.shape(my_sample)[0]
print("Size of the sample:", n)

# Histogram
plt.hist(my_sample, density=True, bins=100)
```

Using the approximation of the survival function
$$\mathbb{P}(T>t) \simeq \frac{1}{n}\sum_{i=1}^n \mathbf{1}_{\{T_i>t\}}$$
complete the following code so that the function `survival(t)` returns $\mathbb{P}(T>t)$, and the function `hazard(t,h)` returns $h^{-1}\mathbb{P}(T \in (t,t+h] | T > t)$.

```python
def survival(t):
  # Your code here
  return np.sum(my_sample>t)/n
```

```python
def hazard(t,h):
  # Your code here
  survival_t = survival(t)
  survival_th = survival(t+h)
  return (survival_t-survival_th)/(h*survival_t)
```

We may now plot the survival function and the hazard rate, for a sufficiently small value of $h$. You may try to observe and understand what happens when you vary the value of $h$ by changing `n_t`.

```python
# Grid of values of t
n_t = 100
t_max = 3 #np.max(my_sample)
h = t_max/n_t
vect_of_t = np.arange(n_t)*h

# Survival function
plt.plot(vect_of_t, [survival(t) for t in vect_of_t], label='Survival function')
plt.legend()
plt.show()

# Hazard rate
plt.plot(vect_of_t, [hazard(t,h) for t in vect_of_t], label='Hazard rate')
plt.legend()
plt.show()
```
